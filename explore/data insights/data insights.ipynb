```python
def open_data(data, n):

    for data_name in data:
        
        df = pd.read_csv(data_name+".csv")
        
        #Remove nan and reset index
        total_len = len(df)
        total_na = df.isna().sum()
        
        df.dropna(inplace=True)
        df.reset_index(drop=True, inplace=True)
        total_na_drop = len(df)
        
        #Removing special characters
        df['message1'] = df.message.apply(remove_string)
        
        #count word in each rows
        df['totalwords'] = [len(x.split()) for x in df['message1'].tolist()]
        
        #Distribution
        label_1 = list(df.label.value_counts())[0]
        label_2 = list(df.label.value_counts())[1]
        #Counting word occurance
        wc = word_freq(df['message1'])
       
        #Sorting words and list
        wcsort = {k: v for k, v in sorted(wc.items(), key=lambda item: item[1], reverse=False)}
        wcsortl = list(wcsort.items())
        
        #Display min and max length of words in data set
        min_words, max_words = min_max(df['message1'])
        
        #Display top most and lest frequent words
        least = wcsortl[:n]
        most = wcsortl[-n:]
        
        print("len(df): {}, len(na): {}, len(na_drop): {}, label_1: {}, label_2:{}, total words: {}, unique words: {}, lest: {}, most: {}, \
              min: {}, max: {}" .format(total_len, total_na, total_na_drop, label_1, label_2, sum(wcsort.values()), len(wc), least, most, 
                                        min_words, max_words))
        #print(total_len, total_na, total_na_drop, sum(wcsortl.values()), len(wc), least, most)
        print("\n")
        
        return df
```

```python
def min_max(documents):
    print(len(documents))
    min = 0
    max = 0
    for doc in documents:
        # donot consider empty document @20ng_CG_RM
#         if len(doc) > 10:
#             if len(doc) < min:
#                 min = len(doc)
        if len(doc) < min:
            min = len(doc)

        if len(doc) > max:
            max = len(doc)
    return min, max
```

```python
def word_freq(documents):
    d = {}
    for i in range(len(documents)):
        for word in documents[i].split():
            try:
                if type(word) != str:
                    word = str(word)
                    
                if word in d:
                    count = d[word] + 1
                    d.update( {word : count} )
                else:
                    d.update( {word : 1} )
            except KeyError:
                print("keyerror")
                print(word, i)
   #print(d)
    print("\n\n\n")
    return d
```
